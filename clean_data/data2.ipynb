# DỮ LIỆU COVID19 THẾ GIỚI 

```python
world_data=pd.read_csv(r'A:\UET-VNU\project\world_data.csv')
```

```python
missing_data_world = world_data.isnull().mean() * 100
missing_data_world
```

## Xử lý các cột của world_data do tỉ lệ thiếu thấp nhỏ hơn 1% tiến hành điền giá trị min và xóa các cột thiếu hơn 10%

```python

missing_value_threshold = 10

# Remove columns with high missing values
world_data = world_data.drop(columns=missing_data_world [missing_data_world > missing_value_threshold].index)

# Impute missing values using median imputation
columns_to_impute = ['total_cases', 'new_cases', 'new_cases_smoothed', 'total_deaths', 'new_deaths', 'new_deaths.1',
                     'total_cases_per_million', 'new_cases_per_million', 'new_cases_smoothed_per_million',
                     'total_deaths_per_million', 'new_deaths_per_million', 'new_deaths_smoothed_per_million',
                     'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations',
                     'new_vaccinations_smoothed', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
                     'people_fully_vaccinated_per_hundred', 'new_people_vaccinated_smoothed',
                     'new_people_vaccinated_smoothed_per_hundred', 'hospital_beds_per_thousand']

world_data[columns_to_impute] = world_data[columns_to_impute].fillna(world_data[columns_to_impute].min())
```

# DỮ LIỆU CHÂU LỤC

```python
contiment_data=pd.read_csv(r'A:\UET-VNU\project\contimets_data.csv')
```

```python
missing_data_contiment = contiment_data.isnull().mean() * 100
missing_data_contiment
```

### Xử lý các cột có tỷ lệ thiếu 100%

continent: 100% thiếu.
hospital_beds_per_thousand: 100% thiếu.
Giải pháp:

Xóa các cột này vì không chứa bất kỳ dữ liệu hữu ích nào.

```python
contiment_data = contiment_data.drop(columns=['continent', 'hospital_beds_per_thousand'])
```

## Xử lý các cột có tỷ lệ thiếu thấp (dưới 10%)
Ví dụ:
Các cột liên quan đến tiêm chủng:
total_vaccinations: 5.30%.
people_vaccinated: 5.30%.
people_fully_vaccinated: 5.91%.
new_vaccinations: 5.45%.
new_vaccinations_smoothed   5.33%
new_people_vaccinated_smoothed 5.33%
new_people_vaccinated_smoothed_per_hundred 5.33%


Hướng xử lý:
Điền giá trị thay thế:
Điền giá trị median cho các cột numeric.

```python
# Điền giá trị trung vị (median) cho các cột numeric
cols_to_fill_median = [
    'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
    'new_vaccinations', 'total_vaccinations_per_hundred',
    'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred','new_vaccinations_smoothed',
    'new_people_vaccinated_smoothed','new_people_vaccinated_smoothed_per_hundred'
    
]
for col in cols_to_fill_median:
    contiment_data[col].fillna(contiment_data[col].median(), inplace=True)


```

## Xử lý các cột có tỷ lệ thiếu từ 0.2% - 1%

Các cột liên quan đến ca bệnh và tử vong:
total_cases, new_cases, total_deaths, new_deaths: 0.27%.
new_cases_smoothed: 0.57%.
new_cases_per_million, total_deaths_per_million: 0.27%.
Hướng xử lý:
Do tỷ lệ thiếu rất thấp, có thể:
Xóa các hàng nếu cần độ chính xác cao.

```python
# loc 
cols_to_fill_mean = [
    'total_cases', 'new_cases', 'total_deaths', 'new_deaths',
    'new_cases_smoothed', 'total_cases_per_million',
    'new_cases_per_million', 'total_deaths_per_million'
]
contiment_data = contiment_data.dropna(subset=cols_to_fill_mean)
```

## Xử lý Cột có tỷ lệ thiếu thấp (dưới 5%)
Cột:
Các cột liên quan đến ca bệnh và tử vong:

Hướng xử lý:
Điền giá trị thay thế:
Vì dữ liệu này liên quan đến số lượng hoặc tỷ lệ, sử dụng giá trị trung bình (mean) hoặc trung vị (median) là phù hợp.


```python
# Điền giá trị trung vị
cols_low_missing = [
    'total_cases', 'new_cases', 'new_cases_smoothed', 'total_deaths', 
    'new_deaths', 'new_deaths.1', 'total_cases_per_million', 
    'new_cases_per_million', 'total_deaths_per_million', 
    'new_deaths_per_million', 'new_deaths_smoothed_per_million'
]
for col in cols_low_missing:
    country_data[col].fillna(country_data[col].median(), inplace=True)
```

## Xử lý cột có tỷ lệ thiếu trung bình (30% - 80%)
Cột:
Các cột liên quan đến tiêm chủng:
total_vaccinations, people_vaccinated, people_fully_vaccinated.
new_vaccinations, new_vaccinations_smoothed.
total_vaccinations_per_hundred, people_vaccinated_per_hundred, people_fully_vaccinated_per_hundred.
new_people_vaccinated_smoothed, new_people_vaccinated_smoothed_per_hundred.hospital_beds_per_thousand
Hướng xử lý:
Điền giá trị thay thế:

Nếu dữ liệu bị thiếu ở khoảng thời gian dài, cân nhắc bỏ các hàng/cột vì không có đủ thông tin.


```python
# Điền giá trị trung vị
cols_vaccinations = [
    'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 
    'new_vaccinations', 'new_vaccinations_smoothed', 
    'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 
    'people_fully_vaccinated_per_hundred', 'new_people_vaccinated_smoothed', 
    'new_people_vaccinated_smoothed_per_hundred','hospital_beds_per_thousand'
]
for col in cols_vaccinations:
    country_data[col].fillna(country_data[col].median(), inplace=True)

```
