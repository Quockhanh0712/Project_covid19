# TỔ CHỨC LẠI DỮ LIỆU

```python
import pandas as pd
```

```python
url=r'A:\UET-VNU\project\covid19_data.csv' # thay đổi đường dẫn tuyệt đối khi chạy
df=pd.read_csv(url)
```

## Điền dữ liệu cho các cột liên quan đến tiêm chủng giá trị 0 từ trước '08/12/2020 

```python

df['date'] = pd.to_datetime(df['date'])
filtered_df = df[df['date'] < pd.to_datetime('2020-12-08')]
columns_to_fill = ['total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'total_boosters', 'new_vaccinations', 'new_vaccinations_smoothed', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred', 
                    'new_vaccinations_smoothed_per_million', 'new_people_vaccinated_smoothed', 'new_people_vaccinated_smoothed_per_hundred']
filtered_df[columns_to_fill] = filtered_df[columns_to_fill].fillna(0)
df.update(filtered_df)
```

```python
# Kiểm tra tỷ lệ dữ liệu thiếu
missing_data = df.isnull().mean() * 100

miss = 75
columns_to_drop = missing_data[missing_data > miss].index
df = df.drop(columns=columns_to_drop)

print(f"Các cột đã bị loại bỏ: {columns_to_drop}")
missing_data
```

# Giữ lại các cột có ý nghĩa cần phân tích

```python


# Tạo một danh sách chứa tên các cột mong muốn
columns_to_select = ['iso_code', 'continent', 'location', 'date', 'total_cases', 'new_cases', 'new_cases_smoothed',
                       'total_deaths', 'new_deaths', 'new_deaths', 'total_cases_per_million', 'new_cases_per_million',
                       'new_cases_smoothed_per_million', 'total_deaths_per_million', 'new_deaths_per_million',
                       'new_deaths_smoothed_per_million', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
                       'new_vaccinations', 'new_vaccinations_smoothed', 'total_vaccinations_per_hundred',
                       'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'new_people_vaccinated_smoothed',
                       'new_people_vaccinated_smoothed_per_hundred', 'hospital_beds_per_thousand']

# Sử dụng danh sách này để chọn các cột từ DataFrame
selected_df = df[columns_to_select]
```

# Chia nhỏ dữ liệu thành các tệp con để xử lý cho phù hợp 


```python
world_data=selected_df[selected_df['location']=='World']
countries_df = selected_df[~selected_df['location'].isin([
    'World',
    'Asia', 'Europe', 'Africa', 'Oceania', 'North America', 'South America',
    'European Union (27)',
    'High-income countries',
    'Upper-middle-income countries',
    'Lower-middle-income countries',
    'Low-income countries'
])]
income_groups = ['High-income countries', 'Upper-middle-income countries', 'Lower-middle-income countries', 'Low-income countries']
income_data=selected_df[selected_df['location'].isin(income_groups)]
contiment=['Asia', 'Europe', 'Africa', 'Oceania', 'North America', 'South America']
contiment_data = selected_df[(selected_df['location'].isin(contiment)) & (selected_df['continent'].isnull())]
contiment_data
```

```python
world_data.to_csv('world_data.csv')
countries_df.to_csv('contries_data.csv')
income_data.to_csv('income_data.csv')
contiment_data.to_csv('contimets_data.csv')
```
